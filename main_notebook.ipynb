# Meow Facts

```python
base_url = "https://meowfacts.herokuapp.com/"
facts_endpoint = "?count=50"
```

```python
# Make a GET request to retrieve facts
response = requests.get(base_url + facts_endpoint)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Parse the JSON response
    meow_facts = response.json()
    
    # Access the facts and create a DataFrame
    facts_list = meow_facts['data']
    meow_facts_df = pd.DataFrame(facts_list, columns=["Fact"])
    display(meow_facts_df)
    
else:
    # Print an error message if the request was not successful
    print(f"Error: {response.status_code}")
```

## Spoonacular

```python
apikey = getpass.getpass()
```

```python
base_url = "https://api.spoonacular.com/recipes/complexSearch"
endpoint_pasta = "?query=pasta"
api_key_param = "&apiKey=" + apikey
```

```python
# Make a GET request to retrieve facts
response = requests.get(base_url + endpoint_pasta + api_key_param)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Parse the JSON response
    recipes = response.json()
    
    # Access the facts and create a DataFrame
    recipes_list = recipes.get('results', [])    
    recipes_df = pd.DataFrame(recipes_list)
    
    pd.set_option('max_colwidth', None)

    display(recipes_df)
    
else:
    # Print an error message if the request was not successful
    print(f"Error: {response.status_code}")
```

## SWAPI

```python
base_url = "https://swapi.dev/api/"
endpoint = "people/1/"
```

```python
# Make a GET request to retrieve facts
response = requests.get(base_url + endpoint)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Parse the JSON response
    response_json = response.json()
    
    # Create a DataFrame from the entire JSON response
    response_df = pd.json_normalize(response_json)

    pd.set_option('max_colwidth', None)

    display(response_df)
    
else:
    # Print an error message if the request was not successful
    print(f"Error: {response.status_code}")
```

```python
base_url = "https://swapi.dev/api/"
endpoint = "people/"

page = 1

all_people = []

while True:
    # Make a GET request to retrieve facts
    response = requests.get(f"{base_url}{endpoint}?page={page}")

    # Check if the request was successful (status code 200)
    if response.status_code == 200:
        # Parse the JSON response
        response_json = response.json()

        # Append the results to the list
        all_people.extend(response_json.get("results", []))

        # Check if there are more pages
        if response_json.get("next"):
            page += 1
        else:
            break
    else:
        # Print an error message if the request was not successful
        print(f"Error: {response.status_code}")
        break

# Create a DataFrame from the list of all people
people_df = pd.DataFrame(all_people)

pd.set_option('max_colwidth', None)
display(people_df)
```
